```python
def search_to_df(return_type,query):
    

    # Create search object
    search = fw.search({'return_type': return_type, 'structured_query': query},size=10000)
    # If no results returned, exit
    if len(search) == 0:
        print('No results returned for search query: {}'.format(query))
        return
    # Initialize data frame list
    df_list = []
    # Iterate over every search item
    for item in search:
        # Set data view object based on return type
        if return_type == 'file':
            return_id = item.parent.id
            df_column = 'acquisition.id'
            view_builder = flywheel.ViewBuilder(columns=['file.name','file.type','project.label'],
                                        container='acquisition', include_ids=True,
                                        include_labels=True, process_files=False,filename=item.file.name)
            # Build the view
            view = view_builder.build()
        elif return_type == 'acquisition':
            view = fw.View(columns=['acquisition.timestamp','acquisition.label'])
            return_id = item[return_type].id
            df_column = return_type + '.id'
        elif return_type == 'session':
            view = fw.View(columns=['session.timestamp','subject.label','project.label'])
            return_id = item[return_type].id
            df_column = return_type + '.id'
        else:
            print('Invalid return_type: {}'.format(return_type))
            os.sys.exit(1)
        project_id = item.project.id
        tmp_df = fw.read_view_dataframe(view, project_id)
    
        tmp_df = tmp_df[tmp_df[df_column]==return_id]
        if len(tmp_df) == 0:
            print("df empty for return id: {}".format(return_id))
        df_list.append(tmp_df)
    df = pd.concat(df_list,ignore_index=True)   
    return df
        
```
